```python
from fastai.vision.image import *

# Assuming your image folder path is stored in 'Dataset_path'
dls = ImageDataLoaders.from_folder(
    path=Dataset_path,
    item_tfms=Resize(224),  # Resize images to 224x224
    bs=16,  # Batch size of 16
    batch_tfms=[Normalize.from_stats(*imagenet_stats), RandTransform()],  # Batch transformations
    valid_pct=0.2  # Validation set percentage
)
```

```python
dls.show_batch()
```

```python
model=vision_learner(
    dls=dls,
    arch=models.resnet50,
    metrics=[accuracy,error_rate]
)
```

```python
model.fit(1)
```
